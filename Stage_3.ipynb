**将以上代码改写为循环体**

```python
from sklearn.decomposition import PCA
#实例化PCA
pca = PCA()
```

```python
for i in range(24,len(yrmn)):
    #切出用于降维的窗口数据
    which_month = yrmn[i]
    start,end = yrmn[i-24],yrmn[i-1]
    window = Before_PCA_std[start:end]
    #训练数据
    pca.fit(window)
    #PCA后的数据
    After_PCA = pd.DataFrame(pca.fit_transform(window),index = window.index,columns = None)
    #计算各自的方差解释度
    ratio = pca.explained_variance_ratio_
    #计算累计方差解释度
    cum_ratio = ratio.cumsum()
    #切出达到累计方差解释度85%的主成分序列
    index_ = int(min(np.argwhere(cum_ratio > 0.85)))
    After_PCA_new = After_PCA.iloc[:,:index_+1]
    #切出这些主成分序列的权重
    weight = ratio[0:index_+1]
    #计算综合情绪指数
    fpc = []
    for i in range(len(After_PCA_new)):
        After_PCA_nums = After_PCA_new.iloc[[i],:]
        fpc += [float(np.dot(After_PCA_nums,weight))]
    After_PCA_new['综合情绪指数'] = fpc
    #切出综合情绪指数最末尾与i月长度相同的一段,作为i月的综合情绪指数
    len_month = len(check_month_len[which_month])
    slicing = After_PCA_new.iloc[-len_month:,-1]
    sentiment['综合情绪指数'][which_month] = slicing

sentiment
```

### 全部A股个股收益率 R_i,t

```python
close_2013 = pd.read_excel('收盘价.xlsx',sheet_name = '2013',index_col = 0).iloc[-1,:]
close_2013 = pd.DataFrame(close_2013).T
close_2014 = pd.read_excel('收盘价.xlsx',sheet_name = '2014',index_col = 0)
close_2014 = close_2013.append(close_2014)
for i in range(len(close_2014.columns)):
    close_2014.iloc[:,i] = pd.to_numeric(close_2014.iloc[:,i],errors='coerce')
close_2014
```

```python
R_it_2014 = pd.DataFrame(index = close_2014.index[1:])
for i in list(close_2014.columns):
    R_it_2014[i] = np.diff(np.log(close_2014[i]))
R_it_2014
```

```python
def R_it(year):
    '''The parameter "year" passes an integer'''
    close_last_year = pd.read_excel('收盘价.xlsx',sheet_name = str(year-1),index_col = 0).iloc[-1,:]
    close_last_year = pd.DataFrame(close_last_year).T
    close_year = pd.read_excel('收盘价.xlsx',sheet_name = str(year),index_col = 0)
    close_year = close_last_year.append(close_year)
    for i in range(4375):
        close_year.iloc[:,i] = pd.to_numeric(close_year.iloc[:,i],errors='coerce')
    R_it_year = pd.DataFrame(index = close_year.index[1:])
    for i in list(close_year.columns):
        R_it_year[i] = np.diff(np.log(close_year[i]))
    return R_it_year

#R_it_2014 = R_it(2014)
R_it_2015 = R_it(2015)
R_it_2016 = R_it(2016)
R_it_2017 = R_it(2017)
R_it_2018 = R_it(2018)
R_it_2019 = R_it(2019)
R_it_2020 = R_it(2020)
```

```python
R_it = pd.concat([R_it_2014,R_it_2015,R_it_2016,R_it_2017,R_it_2018,R_it_2019,R_it_2020])
R_it
```

```python
#将R_it存入excel
R_it.to_excel('R_it.xlsx',index = True)
```

```python
def yrmn(start,end):
    '''produce a list of "yyyy-mm" format'''
    dates = pd.date_range(start=start, end=end, freq="MS")
    yrmn = []
    for i in dates:
        yrmn += [str(i)[:7]]
    return yrmn

yrmn1 = yrmn('2014-01','2020-12')
yrmn1
```

```python
R_mkt_beta_table = pd.DataFrame(index = yrmn('2015-01','2020-12'),columns = R_it.columns)
sentiment_beta_table = pd.DataFrame(index = yrmn('2015-01','2020-12'),columns = R_it.columns)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

def betas_per_stock(stock):
    '''For "stock" [an integer in range(4375)], return R_mkt_beta & sentiment_beta for 12-month \
    period [including dropped period] before each month since Jan 2015'''
    # 提取个股用于回归的数据、设置放置回归系数的list
    R_mkt_beta = []
    sentiment_beta = []
    stock_ = pd.concat([R_it.iloc[:,stock],R_mkt.iloc[:,0],sentiment.iloc[:,0]],axis = 1)#以行对齐
    # 提取个股交易状态不合格的数据
    allok_stock = pd.DataFrame(allok.iloc[stock,:])
    allok_stock.index = pd.to_datetime(allok_stock.index) # pandas转换series至dataframe时间标签                                            
    col_name = list(allok_stock.columns)[0]               # '2020-12'出错，需要pd.to_datetime修改
    allok_stock = allok_stock[allok_stock[col_name] == 0]
    filter_stock = []
    for i in list(allok_stock.index):
        filter_stock += [str(i.year) + '-' + str(i.month)]
    # 过滤个股用于回归的数据(不合格的交易状态)
    for j in set(filter_stock):                          # pandas转换series至dataframe时间标签
        index_ = list(stock_[j].index)                   # '2020-12'重复，需要set修改
        stock_ = stock_.drop(index = index_)
    # 过滤个股用于回归的数据(最初源自收盘价空白值，导入后变成nan)
    index_1 = list(stock_.index[np.where(np.isnan(stock_))[0]])
    stock_ = stock_.drop(index = index_1)    
    # 逐月回归,每月得到2个回归系数，得到两列回归系数
    for k in range(12,len(yrmn1)):
        start,end = yrmn1[k-12],yrmn1[k-1]
        slicing = stock_[start:end]
        if len(slicing.index) >= 1:         
            lr.fit(slicing.iloc[:,[1,2]],slicing.iloc[:,0])
            R_mkt_beta += [round(list(lr.coef_)[0], 6)]
            sentiment_beta += [round(list(lr.coef_)[1], 11)]
        else:
            R_mkt_beta += [np.nan]
            sentiment_beta += [np.nan]
    # 将两列回归系数放入最终表格
    R_mkt_beta_table.iloc[:,stock] = R_mkt_beta
    sentiment_beta_table.iloc[:,stock] = sentiment_beta
```

```python
for stock_i in range(len(R_it.columns)):
    betas_per_stock(stock_i)
```
