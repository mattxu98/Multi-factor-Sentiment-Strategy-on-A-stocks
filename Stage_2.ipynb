### 3 通过主成分分析(PCA)构建情绪指数，阈值（方差解释度）为85%

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd

Before_PCA = pd.read_excel('Before_PCA.xlsx',index_col = 0)
Before_PCA
```

```python
#对x做标准化处理
from sklearn.preprocessing import StandardScaler
Before_PCA_std = StandardScaler().fit_transform(Before_PCA)
Before_PCA_std = pd.DataFrame(Before_PCA_std,index = Before_PCA.index,columns = Before_PCA.columns)
Before_PCA_std
```

```python
from sklearn.decomposition import PCA
#实例化PCA
pca = PCA()
#训练模型
pca.fit(Before_PCA_std)
```

```python
#PCA后的数据
After_PCA = pd.DataFrame(pca.fit_transform(Before_PCA_std),index = Before_PCA.index,columns = None)
After_PCA
```

```python
#计算各自的方差解释度
ratio = pca.explained_variance_ratio_
ratio
```

```python
#计算累计方差解释度
cum_ratio = ratio.cumsum()
cum_ratio
```

**如以上结果所示，前4个指标的累计方差解释度满足85%的要求，故n_components = 4**

```python
#调参，重新实例化PCA
pca1 = PCA(n_components = 4)
#训练模型
pca1.fit(Before_PCA_std)
```

```python
#对Before_PCA_std进行降维
After_PCA_new = pca1.transform(Before_PCA_std)
After_PCA_new = pd.DataFrame(After_PCA_new,index = Before_PCA.index,columns = None)
After_PCA_new.columns = ['first PC','second PC','third PC','fourth PC']
After_PCA_new
```

```python
# 权重
weight=pca1.explained_variance_ratio_
print(weight)
print(sum(weight))
```

#### 计算综合情绪指数

**注意到以上第一列weight只有0.50074351，采用加权求和**

```python
import numpy as np
#计算第一主成分序列
fpc = []
for i in range(len(After_PCA_new)):
    After_PCA_nums = After_PCA_new.iloc[[i],:]
    fpc += [float(np.dot(After_PCA_nums,weight))]
After_PCA_new['综合情绪指数'] = fpc
After_PCA_new
```
